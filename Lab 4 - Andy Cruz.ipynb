```python
import pandas as pd
data_1 = pd.read_csv("/content/PHYS471/LASER_MEASURE.csv",  names=['aperture_r', 'volt_1','volt_2','volt_3','volt_4','volt_5'])
```

Intensity vs Aperture Radius

```python
import matplotlib.pyplot as plt
plt.scatter(data_1.aperture_r,data_1.volt_1)
plt.scatter(data_1.aperture_r,data_1.volt_3)
plt.scatter(data_1.aperture_r,data_1.volt_4)
plt.scatter(data_1.aperture_r,data_1.volt_5)
plt.xlabel('Aperture Radius (mm)')
plt.ylabel('Intensity(V)')
plt.title('Sensor Voltage Against Aperture Radius')
plt.show()
```

Intensity vs Aperture Area

```python
import numpy as np
aperture_A = np.pi*data_1.aperture_r**2
plt.scatter(aperture_A,data_1.volt_1,label = 'Cyclic RMS 2.03 V')
plt.scatter(aperture_A,data_1.volt_3,label = 'Cyclic RMS 1.78 V Run 1')
plt.scatter(aperture_A,data_1.volt_4,label = 'Cyclic RMS 1.78 V Run 2')
plt.scatter(aperture_A,data_1.volt_5,label = 'Cyclic RMS 1.57 V')
plt.xlabel('Aperture Area (mm^2)')
plt.ylabel('Intensity (V)')
plt.title('Intensity Against Aperture Area')
plt.legend()
plt.show()
```
