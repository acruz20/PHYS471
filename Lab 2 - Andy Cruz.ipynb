```python
import pandas as pd
data_1 = pd.read_csv("/content/PHYS471/ROD_PIEZO_1.csv", usecols=[3,4,9,10], names=['time_A', 'volt_A','time_B','volt_B'])
data_2 = pd.read_csv("/content/PHYS471/ROD_PIEZO_2.csv", usecols=[3,4,9,10], names=['time_A', 'volt_A','time_B','volt_B'])
data_3 = pd.read_csv("/content/PHYS471/ROD_PIEZO_3.csv", usecols=[3,4,9,10], names=['time_A', 'volt_A','time_B','volt_B'])
data_4 = pd.read_csv("/content/PHYS471/ROD_PIEZO_4.csv", usecols=[3,4,9,10], names=['time_A', 'volt_A','time_B','volt_B'])
data_5 = pd.read_csv("/content/PHYS471/ROD_PIEZO_5.csv", usecols=[3,4,9,10], names=['time_A', 'volt_A','time_B','volt_B'])
```

```python
def edge_detA(volt,time):
  trigger = 0
  n = 1
  while trigger<=0.1:
    n += 1
    trigger = (volt[n-2]+volt[n-1]+volt[n]+volt[n+1]+volt[n+2])/5
  return time[n]

def edge_detB(volt,time):
  trigger = 0
  n = 1
  while trigger<=0.5:
    n += 1
    trigger = (volt[n-2]+volt[n-1]+volt[n]+volt[n+1]+volt[n+2])/5
  return time[n]

del_1 = edge_detA(data_1.volt_A,data_1.time_A) - edge_detB(data_1.volt_B,data_1.time_B)
del_2 = edge_detA(data_2.volt_A,data_2.time_A) - edge_detB(data_2.volt_B,data_2.time_B)
del_3 = edge_detA(data_3.volt_A,data_3.time_A) - edge_detB(data_3.volt_B,data_3.time_B)
del_4 = edge_detA(data_4.volt_A,data_4.time_A) - edge_detB(data_4.volt_B,data_4.time_B)
del_5 = edge_detA(data_5.volt_A,data_4.time_A) - edge_detB(data_5.volt_B,data_5.time_B)
print(del_1,del_2,del_3,del_4,del_5)

del_t = (del_1+del_2+del_3+del_4+del_5)/5
rod_l = 0.9095
speed = rod_l/del_t
print(speed)
```

3178 m/s is the calculated speed from the averages of the time differences

```python
import matplotlib.pyplot as plt
#plt.scatter([data_1.time_A,data_1.time_B],[data_1.volt_A,data_1.volt_B])
plt.scatter(data_1.time_A,data_1.volt_A, label = 'Piezo Voltage')
plt.axvline(x = edge_detA(data_1.volt_A,data_1.time_A), color = 'blue', label = 'Initial Piezo Contact Time')
plt.scatter(data_1.time_B,data_1.volt_B, label = 'Rod Voltage')
plt.axvline(x = edge_detB(data_1.volt_B,data_1.time_B), color = 'orange', label = 'Initial Hammer Contact Time')
plt.xlabel('Relative Time from Hammer Liftoff (s)')
plt.ylabel('Voltage (V)')
plt.title('Impact Times for Hammer and Piezo Run 1')
plt.legend()
plt.show()
```
